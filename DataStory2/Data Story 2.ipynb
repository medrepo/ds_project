```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

%matplotlib inline
```

```python
import json
from pandas.io.json import json_normalize

# Retrieve dataset from source
dfhomesale = pd.read_json('https://www.quandl.com/api/v3/datasets/NAR/HOME_SALES_ALL.json?collapse=annual')
```

```python
# Parse dataset into a dataframe with correct format for year column
data = pd.DataFrame(dfhomesale['dataset'][3], columns=(dfhomesale['dataset'][2]))
data['Year'] = data['Year'].astype('datetime64')

usabledf = data[['Year','Single-Family', 'Condo/Co-op']]
def percchange(salecurr, salelastyr):
    pchange = np.round(((salecurr - salelastyr) / salelastyr) * 100, 2)
    
    return pchange

realdf = pd.DataFrame(columns=['year','Single-Family','Condo'], index=range(len(usabledf) - 1))
realdf['year'] = (usabledf['Year'].dt.year).astype('string')

for x in range(len(usabledf) - 1):
        realdf['Single-Family'][x] = percchange(usabledf['Single-Family'][x], usabledf['Single-Family'][x+1])
        realdf['Condo'][x] = percchange(usabledf['Condo/Co-op'][x], usabledf['Condo/Co-op'][x+1])

realdf = realdf.sort()
realdf = realdf.set_index(['year'])

#realdf.plot(title='percentage change in sales from previous year', marker='*').legend(loc='center left', bbox_to_anchor=(1.05, 0.8));

# plot seperate graphs
fig, axes = plt.subplots(nrows=2, ncols = 1, figsize=(10,10))
realdf['Single-Family'].plot(ax=axes[0], title='percentage change in sales per yr for Single-Family homes', color='g')
realdf['Condo'].plot(ax=axes[1], title='percentage change in sales per yr for Condos/Co-ops', color='b')

# setting the axis labels
for a in range(len(axes)):
    axes[a].set_xlabel('years')
    axes[a].set_ylabel('% of passengers')
plt.tight_layout();
```

Based on the data above, it would be interesting to know which type of home people are investing in more each year on average.

Is the average percentage increase/decrease (change) in sales, per year, for a single family home more than that for a condo? 


* Our Hypothesis:
    
$ H_0: \mu_{sf} - \mu_c = 0$
    
$ H_a: \mu_{sf} - \mu_c > 0$



* Verify conditions for inference:

1) Independence: The data is for only 6 years which is < 10% of the population (all the years through which homes have been sold). The dataset has been randomly picked for these 6 years, hence the observations are independent.

2) Sample size/skew: The sample size is small < 10. Hence we will follow the t-distribution.


* Find the test statistic (T) and make a decision based on the corresponding p-value:


```python
# set the null value and degrees of freedom
nv = 0
df = len(realdf) - 1

# find the sample mean and standard deviation
msf = np.abs(realdf['Single-Family']).mean()
mc = np.abs(realdf['Condo']).mean()

ssf = msf = np.abs(realdf['Single-Family']).std()
sc = np.abs(realdf['Condo']).std()

# find the difference in means and standard error
mdiff = round(msf - mc, 2)
SE = round(np.sqrt(((ssf ** 2)/len(realdf)) + ((sc ** 2)/len(realdf))),2)

# calculate the test statistic
T = round((mdiff - nv) / SE, 2)
p = round(st.t.sf(T, df), 4)

print "The test statistic (T) = %s with degrees of freedom = %s" % (T, df)
print "The one-sided p-value = %s" % p
```

**Based on the p-value above, since it is > $\alpha$ (0.05), we fail to reject the null hypothesis and conclude that the data provides sufficient evidence that the average percentage change (increase or decrease) per year for a Single-Family home is not always more than that for a Condo/Co-op.**

* What would be the difference in average % change between the two?

```python
t = np.abs(round(st.t.ppf(1-0.05, df),2))

me = round(t * SE,2)
nci = round(mdiff - me,2)
pci = round(mdiff + me,2)

print "The confidence interval is (%s, %s)" % (nci, pci)
```

**We are 95% confident that the average percentage change in sales of Single-Family homes would be about 9.3% lower to about 3.82% higher than the percentage change (increase/decrease) in sales of Condos/Co-ops per year.**
